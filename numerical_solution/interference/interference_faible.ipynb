```python
import numpy as np
from scipy.constants import c #contant de Plank et vitesse de la lumière
import matplotlib.pyplot as plt
import fonction_interference as fi
```

```python
wavelength = 640e-9 #nm
largeur = 10e-9 #ns
MIN  = -5*largeur
MAX = 5*largeur
N = 1000
angle_pre_selection = np.pi/4
ellipse_pre = 0
phi_x = 0
phi_y = 0
frequence = c/wavelength
freq_angulaire = 2*np.pi*frequence
d = 0.003 #différence de distance parcouru des deux faisceau (m)
```

```python
t = np.linspace(MIN, MAX, N)
A_t = fi.fonction_gaussien(largeur, t, 0, N)
E = fi.impulsion_gaussien(angle_pre_selection, phi_x, phi_y, freq_angulaire, largeur, t, 0, N)
I = fi.intensity(E)



plt.figure()
plt.plot(t, I)
plt.show()
```

```python
I_1, I_2 = fi.michelson_interference(E, d)

plt.figure()
plt.plot(t, I_1)
plt.plot(t, I_2)
plt.show()
```

```python
tau = d/c
A_t = fi.fonction_gaussien(largeur, t, -tau, N)
E_2 = fi.impulsion_gaussien(angle_pre_selection, phi_x, phi_y, freq_angulaire, largeur, t, -tau, N)
I_2 = fi.intensity(E_2)

plt.figure()
plt.plot(t, I)
plt.plot(t, I_2)
plt.show()
```
